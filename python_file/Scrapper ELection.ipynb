```python
from bs4 import BeautifulSoup
```

```python
def extract_info(soup,Constituency,Parishad, two_class): 
    arrr = []
    trrr = soup.find("tbody", class_ = "ng-scope").find_all("tr", class_ = two_class)
    for i in range(len(trrr)):
        try: 
            # S.no
            s_no = int(trrr[i].find_all("td")[0].text) 
            # Name 
            cand_name = trrr[i].find_all("td")[1].find('span').text
            # Sex
            sex = trrr[i].find_all("td")[2].text.strip()
            #Address
            Address = trrr[i].find_all("td")[3].text.strip()
            # Party Affiliation
            party = trrr[i].find_all("td")[4].text.strip()
            # Category
            category = trrr[i].find_all("td")[5].text.strip()

            eachrow = {"S_no": s_no, "Constituency":Constituency, "Parishad":Parishad, "Name": cand_name, "Gender": sex, "Address": Address, "Party_Affiliation": party, "Category": category}     
            arrr.append(eachrow)
        except:
            pass
    return arrr
```
